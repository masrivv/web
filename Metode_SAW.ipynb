```python
def input_data():
    # Input jumlah alternatif dan kriteria
    A = int(input("Masukkan jumlah alternatif: "))
    C = int(input("Masukkan jumlah kriteria: "))

    # Input jenis kriteria (benefit atau cost)
    kriteria = []
    for i in range(C):
        jenis = input(f"Masukkan jenis kriteria C{i + 1} (benefit/cost): ")
        kriteria.append(jenis)

    # Input nilai bobot setiap kriteria
    bobot = []
    for i in range(C):
        b = float(input(f"Masukkan nilai bobot kriteria C{i + 1}: "))
        bobot.append(b)

    # Input nilai setiap kriteria untuk setiap alternatif
    matriks = []
    for i in range(A):
        nilai_alternatif = []
        print(f"Masukkan nilai untuk alternatif A{i + 1}:")
        for j in range(C):
            nilai = float(input(f"Kriteria ke-{j + 1}: "))
            nilai_alternatif.append(nilai)
        matriks.append(nilai_alternatif)

    return A, C, kriteria, bobot, matriks


def normalisasi_matriks(matriks, kriteria, C):
    matriks_ternormalisasi = []
    for j in range(C):
        kolom = [matriks[i][j] for i in range(len(matriks))]
        if kriteria[j] == 'benefit':
            norm = [x / max(kolom) for x in kolom]
        elif kriteria[j] == 'cost':
            norm = [min(kolom) / x for x in kolom]
        for i in range(len(matriks)):
            if j == 0:
                matriks_ternormalisasi.append([norm[i]])
            else:
                matriks_ternormalisasi[i].append(norm[i])

    return matriks_ternormalisasi


def hitung_skala_prioritas(matriks_ternormalisasi, bobot):
    skala_prioritas = []
    for i in range(len(matriks_ternormalisasi)):
        total = sum(matriks_ternormalisasi[i][j] * bobot[j] for j in range(len(bobot)))
        skala_prioritas.append(total)
    return skala_prioritas


def main():
    A, C, kriteria, bobot, matriks = input_data()
    matriks_ternormalisasi = normalisasi_matriks(matriks, kriteria, C)
    skala_prioritas = hitung_skala_prioritas(matriks_ternormalisasi, bobot)

    print("\nHasil:")
    for i in range(A):
        print(f"Alternatif ke-{i + 1} : {skala_prioritas[i]}")

    # Menampilkan alternatif dengan skor tertinggi sebagai alternatif terbaik
    terbaik = skala_prioritas.index(max(skala_prioritas)) + 1
    print(f"\nAlternatif terbaik adalah Alternatif ke-{terbaik}")


if __name__ == "__main__":
    main()
```
